#### Import modules :

```python
# import modules
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

#### load dataset from :

```python
# print the iris data
iris = load_iris()
x = iris.data
y = iris.target
print (iris.target_names)
```

#### Calculate the Cosine Similarity:
The Cosine similarity is a measure of similarity between two non-zero vectors of an inner product space that
measures the cosine of the angle between them:
                        (X, Y ) = < X, Y > /(||X|| âˆ— ||Y||)

```python
iris_cosine_similarity = cosine_similarity(x,x)
# visualize the resultant similarity matrix
im = plt.imshow(iris_cosine_similarity, cmap='RdBu_r') 
plt.colorbar(im, orientation='horizontal')
plt.show()
```

### Plot the histogram for each class :
Here we plot the histogram for each attribute (sepal length, sepal width, petal length, petal width),
and each class with different color. 

```python
fig, axes = plt.subplots(nrows= 2, ncols=2)
colors= ['blue', 'red', 'green']
for i, ax in enumerate(axes.flat):
    for label, color in zip(range(len(iris.target_names)), colors):
        ax.hist(x[iris.target==label, i], label=             
                            iris.target_names[label], color=color)
        ax.set_xlabel(iris.feature_names[i])  
        ax.legend(loc='upper right')
plt.show()
```
